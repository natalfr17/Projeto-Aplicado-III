D) **Simulação de Carteiras via Monte Carlo e Fronteira Eficiente**

Para a simulação de carteiras, foi utilizada a metodologia de Monte Carlo. Foram realizadas 50.000 simulações de carteiras aleatórias para construir a fronteira eficiente. Cada carteira teve seu retorno esperado, volatilidade (risco) e Índice de Sharpe calculados. O gráfico de dispersão no espaço risco-retorno mostra o conjunto de carteiras, onde a fronteira eficiente é a parte superior-esquerda, representando carteiras com o maior retorno para um dado risco ou o menor risco para um dado retorno, obtida por otimização de minimização de volatilidade para diferentes níveis de retorno esperados.

```python
# D) Simular carteiras via Monte Carlo
acoes_port = acoes_df.copy()
acoes_port.drop(labels=['^BVSP'], axis=1, inplace=True)
log_ret = acoes_port.copy()
log_ret.drop(labels=["Date"], axis=1, inplace=True)
log_ret = np.log(log_ret / log_ret.shift(1))

np.random.seed(42)
num_ports = 10000
all_weights = np.zeros((num_ports, len(acoes_port.columns[1:])))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    weights = np.array(np.random.random(len(acoes_port.columns[1:])))
    weights = weights / np.sum(weights)
    all_weights[x, :] = weights
    ret_arr[x] = np.sum((log_ret.mean() * weights))
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov(), weights)))
    sharpe_arr[x] = ret_arr[x] / vol_arr[x]

def get_ret_vol_sr(weights):
    weights = np.array(weights)
    ret = np.sum(log_ret.mean() * weights)
    vol = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov(), weights)))
    sr = ret / vol
    return np.array([ret, vol, sr])

def neg_sharpe(weights):
    return get_ret_vol_sr(weights)[2] * -1

def check_sum(weights):
    return np.sum(weights) - 1

cons = ({'type': 'eq', 'fun': check_sum})
bounds = tuple((0, 1) for x in range(len(acoes_port.columns[1:])))
init_guess = [1. / len(acoes_port.columns[1:])] * len(acoes_port.columns[1:])

# Otimização para a fronteira eficiente
frontier_y = np.linspace(vol_arr.min(), vol_arr.max(), 100) # Ajuste do range para a fronteira
frontier_x = []

def minimize_volatility(weights):
    return get_ret_vol_sr(weights)[1]

for possible_return in np.linspace(ret_arr.min(), ret_arr.max(), 100):
    cons = ({'type': 'eq', 'fun': check_sum},
            {'type': 'eq', 'fun': lambda w: get_ret_vol_sr(w)[0] - possible_return})
    result = optimize.minimize(minimize_volatility, init_guess, method='SLSQP', bounds=bounds, constraints=cons)
    frontier_x.append(result['fun'])


plt.figure(figsize=(8, 4))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatilidade')
plt.ylabel('Retorno')
plt.plot(frontier_x, np.linspace(ret_arr.min(), ret_arr.max(), 100), 'b--', linewidth=3, label='Fronteira Eficiente')
plt.title('Simulação de Carteiras e Fronteira Eficiente')
plt.legend()
plt.show()
```

E) **Identificação da Carteira com Maior Índice de Sharpe**

A carteira com o maior Índice de Sharpe, destacada no gráfico em vermelho, é a carteira ótima, maximizando o retorno por unidade de risco. O Sharpe Ratio máximo encontrado é de 0.014979. Os pesos alocados são: ITUB3.SA (64.99%), MGLU3.SA (26.34%), CIEL3.SA (5.35%), PETR3.SA (1.05%) e CSNA3.SA (2.27%). A racionalidade da composição reflete uma estratégia de buscar estabilidade com ITUB3.SA (menor volatilidade, retorno consistente) e crescimento com MGLU3.SA (potencial de retorno), enquanto minimiza o risco de ativos mais voláteis com pequenas alocações.


```python
max_sr_ret = ret_arr[sharpe_arr.argmax()]
max_sr_vol = vol_arr[sharpe_arr.argmax()]
melhores_pesos = all_weights[sharpe_arr.argmax(), :]

print(f"\nMax Sharpe Ratio: {sharpe_arr.max()}")
print(f"Local do Max Sharpe Ratio: {sharpe_arr.argmax()}")
print(f"Pesos do Portfólio do Max Sharpe Ratio: {melhores_pesos}")
print(f"Retorno da carteira com Max Sharpe Ratio: {max_sr_ret*100:.4f}%")
print(f"Volatilidade da carteira com Max Sharpe Ratio: {max_sr_vol*100:.4f}%")

plt.figure(figsize=(8, 4))
plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatilidade')
plt.ylabel('Retorno')
plt.scatter(max_sr_vol, max_sr_ret, c='red', s=200, edgecolors='black', label='Carteira com Max Sharpe Ratio')
plt.plot(frontier_x, np.linspace(ret_arr.min(), ret_arr.max(), 100), 'b--', linewidth=3, label='Fronteira Eficiente')
plt.title('Simulação de Carteiras e Fronteira Eficiente com Carteira Ótima')
plt.legend()
plt.show()
```

F) **Simulação da Evolução do Patrimônio**

Assumindo um investimento inicial de R$ 35.000,00, a trajetória temporal do patrimônio acumulado com base na carteira ótima (aquela com o maior Sharpe Ratio) foi simulada e visualizada no gráfico "Evolução do patrimônio da Carteira".

O gráfico demonstra claramente a valorização da carteira ao longo do tempo, partindo do investimento inicial e atingindo um valor final de R$ 159.070,07. Essa evolução positiva se reflete no retorno acumulado no período, que foi de aproximadamente 354.49%. A curva de crescimento, apesar de apresentar flutuações esperadas em um mercado de renda variável, indica uma tendência ascendente consistente, validando a eficácia da carteira otimizada em gerar um retorno substancial sobre o capital investido ao longo do período analisado.
# Nova seção

```python
# F) Simular a evolução do patrimônio
dinheiro_total_inicial = 35000.00 # Investimento inicial de R$ 35.000,00

def alocacao_ativos(dataset, dinheiro_total, seed=0, melhores_pesos=[]):
    dataset = dataset.copy()
    if seed != 0:
        np.random.seed(seed)
    if len(melhores_pesos) > 0:
        pesos = melhores_pesos
    else:
        pesos = np.random.random(len(dataset.columns) - 1)
        pesos = pesos / pesos.sum()

    colunas = dataset.columns[1:]
    for i in colunas:
        dataset[i] = (dataset[i] / dataset[i][0])

    for i, acao in enumerate(dataset.columns[1:]):
        dataset[acao] = dataset[acao] * pesos[i] * dinheiro_total

    dataset['soma valor'] = dataset.drop(columns=['Date']).sum(axis=1) # Excluir 'Date' da soma
    datas = dataset['Date']
    dataset.drop(labels=['Date'], axis=1, inplace=True)
    dataset['taxa retorno'] = 0.0
    for i in range(1, len(dataset)):
        dataset['taxa retorno'][i] = np.log(dataset['soma valor'][i] / dataset['soma valor'][i - 1])

    acoes_pesos = pd.DataFrame(data={'Ações': colunas, 'Pesos': pesos})
    return dataset, datas, acoes_pesos, dataset.loc[len(dataset) - 1]['soma valor']

# Usando os melhores pesos encontrados na simulação Monte Carlo
dataset_alocacao, datas_alocacao, acoes_pesos_alocacao, soma_valor_final = alocacao_ativos(acoes_port, dinheiro_total_inicial, 0, melhores_pesos)

print("\nEvolução do Patrimônio (DataFrame detalhado):")
print(dataset_alocacao.head())
print(dataset_alocacao.tail())
print(f"\nValor final do patrimônio: R$ {soma_valor_final:.2f}")

retorno_acumulado = ((soma_valor_final - dinheiro_total_inicial) / dinheiro_total_inicial) * 100
print(f"Retorno acumulado no período: {retorno_acumulado:.2f}%")

largura_desejada = 800
altura_desejada = 500
figura = px.line(x=datas_alocacao, y=dataset_alocacao['soma valor'],
                 title='Evolução do patrimônio da Carteira',
                 labels=dict(x="Data", y="Valor R$"))
figura.add_hline(y=dataset_alocacao['soma valor'].mean(), line_color="green", line_dash="dot", annotation_text=f"Média: R$ {dataset_alocacao['soma valor'].mean():.2f}")
figura.add_hline(y=dinheiro_total_inicial, line_color="red", line_dash="dot", annotation_text=f"Investimento Inicial: R$ {dinheiro_total_inicial:.2f}")
figura.update_layout(width=largura_desejada, height=altura_desejada)
figura.show()

figura_ret_diario = px.line(x=datas_alocacao, y=dataset_alocacao['taxa retorno'], title='Retorno diário do portfólio', labels=dict(x="Data", y="Retorno %"))
figura_ret_diario.add_hline(y=dataset_alocacao['taxa retorno'].mean(), line_color="red", line_dash="dot", annotation_text=f"Média: {dataset_alocacao['taxa retorno'].mean():.4f}%")
figura_ret_diario.update_layout(width=largura_desejada, height=altura_desejada)
figura_ret_diario.show()
```

G) **Avaliação do Risco da Carteira via Value at Risk (VaR)**

O Value at Risk (VaR) estima a perda máxima esperada de um portfólio em um período e nível de confiança. O VaR Histórico é calculado a partir dos retornos diários da carteira.

    VaR Histórico (95% de Confiança):

    Percentual: -0.0317

Monetário: R$ -1109.50

VaR Histórico (99% de Confiança):

    Percentual: -0.0528

Monetário: R$ -1848.00

Interpretação:


Em 95% dos dias, a perda diária da carteira não deve exceder R$ 1109.50. Em 99% dos dias, a perda diária não deve exceder R$ 1848.00. Quanto maior o valor absoluto do VaR, maior o risco de perda para a carteira.

```python
# Retornos diários da carteira ótima
retornos_carteira = dataset_alocacao['taxa retorno'].dropna()

# VaR Histórico (95% de Confiança)
var_95_percentual = np.percentile(retornos_carteira, 5)
var_95_monetario = var_95_percentual * soma_valor_final

# VaR Histórico (99% de Confiança)
var_99_percentual = np.percentile(retornos_carteira, 1)
var_99_monetario = var_99_percentual * soma_valor_final

print(f"\n--- Value at Risk (VaR) da Carteira Ótima ---")
print(f"Investimento Inicial: R$ {dinheiro_total_inicial:.2f}")
print(f"Valor Atual da Carteira: R$ {soma_valor_final:.2f}")

print(f"\nVaR Histórico (95% de Confiança):")
print(f"  Percentual: {var_95_percentual:.4f}")
print(f"  Monetário: R$ {var_95_monetario:.2f}")

print(f"\nVaR Histórico (99% de Confiança):")
print(f"  Percentual: {var_99_percentual:.4f}")
print(f"  Monetário: R$ {var_99_monetario:.2f}")

print("\nInterpretação do Risco Envolvido:")
print(f"Em 95% dos dias, a perda diária da carteira não deve exceder R$ {-var_95_monetario:.2f}.")
print(f"Em 99% dos dias, a perda diária da carteira não deve exceder R$ {-var_99_monetario:.2f}.")
```
